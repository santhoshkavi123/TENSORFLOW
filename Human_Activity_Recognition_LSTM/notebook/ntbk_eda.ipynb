# UDFs

```python
def viz_time_series(X : np.ndarray, y: np.ndarray, index_value: int = None):
    
    # label mapping 
    label_mapping = {
        "0": "WALKING",
        "1": "WALKING_UPSTAIRS",
        "2": "WALKING_DOWNSTAIRS",
        "3": "SITTING",
        "4": "STANDING",
        "5": "LAYING"
    }

    # Random Choice
    if index_value == None:
        random_sample = np.random.randint(0, X.shape[0])
    else:
        random_sample = index_value

    # get label 
    indices = np.where(y[random_sample, :] == 1)
    get_label = label_mapping[str(indices[0][0])]

    legend = ["body_acc_x", "body_acc_y", "body_acc_z", "body_gyro_x", "body_gyro_y", "body_gyro_z", "total_acc_x", "total_acc_y", "total_acc_z"]

    # get sample
    plt.figure(figsize = (20, 15))
    for i in range(0, X.shape[-1]):
        ax = plt.subplot(3, 3, i+1)
        plt.plot(X[random_sample, : , i], label = legend[i])
        plt.xlabel("Readings")
        plt.legend()
        plt.title(str(get_label)+"-"+str(random_sample))
    plt.show()

```
